```python
# Importing essential libraries for data analysis and visualization
import pandas as pd  # For data manipulation and analysis (DataFrames)
import numpy as np   # For numerical computations
import matplotlib.pyplot as plt  # For creating visualizations
import seaborn as sns  # For advanced and aesthetically pleasing visualizations

# Set Seaborn style for plots (white grid for a clean look, like a detective/thriller vibe)
sns.set(style="whitegrid")

# Set Matplotlib style to 'ggplot' for better visuals
plt.style.use("ggplot") 

# Suppress warnings to keep notebook output clean and focused
import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 2: Load Dataset

df = pd.read_csv("creditcard.csv")

# First look at the data
```

```python
# Step 3: Basic Dataset Overview

# Print the number of rows and columns in the dataset
print("Dataset Shape:", df.shape)

# Display information about each column: data type, non-null count, etc.
print("\nColumn Info:")
print(df.info())

# Check for missing values in each column
print("\nMissing Values:\n", df.isnull().sum())

# Show basic statistical summary for numerical columns: count, mean, std, min, max, percentiles
print("\nBasic Statistics:\n", df.describe())
```

```python
# Step 3: Basic Dataset Overview

# Print the number of rows and columns in the dataset
print("Dataset Shape:", df.shape)

# Display information about each column: data type, non-null count, etc.
print("\nColumn Info:")
print(df.info())

# Check for missing values in each column
print("\nMissing Values:\n", df.isnull().sum())

# Show basic statistical summary for numerical columns: count, mean, std, min, max, percentiles
print("\nBasic Statistics:\n", df.describe())
```

```python
# Step 4: Class Distribution

# Create a figure of size 6x4 inches
plt.figure(figsize=(6,4))

# Plot the count of each class (0 = Non-Fraud, 1 = Fraud) using Seaborn's countplot
sns.countplot(data=df, x='Class', palette='coolwarm')

# Add title to the plot
plt.title("Fraud vs Non-Fraud Transactions")

# Rename x-axis ticks for clarity
plt.xticks([0, 1], ['Non-Fraud (0)', 'Fraud (1)'])

# Display the plot
plt.show()

# Separate the dataset into fraud and non-fraud cases
fraud_cases = df[df['Class'] == 1]
nonfraud_cases = df[df['Class'] == 0]

# Calculate the percentage of fraud cases in the dataset
fraud_percent = (len(fraud_cases) / len(df)) * 100

# Print the number and percentage of fraud and non-fraud cases
print(f"Fraud cases: {len(fraud_cases)} ({fraud_percent:.4f}%)")
print(f"Non-Fraud cases: {len(nonfraud_cases)}")
```

```python
# Step 5: Time Feature Analysis

# Plot the distribution of all transactions over time
plt.figure(figsize=(8,4))
sns.histplot(df['Time'], bins=50, kde=False, color='blue')  # Histogram with 50 bins, no KDE
plt.title("Transaction Frequency over Time (seconds)")  # Title of the plot
plt.xlabel("Time (in seconds since first transaction)")   # X-axis label
plt.ylabel("Number of Transactions")                     # Y-axis label
plt.show()

# Plot the distribution of fraud transactions over time
plt.figure(figsize=(8,4))
sns.histplot(fraud_cases['Time'], bins=50, kde=False, color='red')  # Histogram for fraud cases
plt.title("Fraud Transaction Frequency over Time")  # Title for fraud plot
plt.xlabel("Time (in seconds)")                     # X-axis label
plt.ylabel("Number of Fraud Transactions")          # Y-axis label
plt.show()
```

```python
# Step 6: Amount Feature Analysis

# Plot the distribution of all transaction amounts
plt.figure(figsize=(8,4))
sns.histplot(df['Amount'], bins=50, color='green')  # Histogram with 50 bins for Amount
plt.title("Transaction Amount Distribution")        # Title for the plot
plt.xlabel("Amount ()")                             # X-axis label
plt.ylabel("Number of Transactions")                # Y-axis label
plt.show()

# Plot the distribution of fraud transaction amounts
plt.figure(figsize=(8,4))
sns.histplot(fraud_cases['Amount'], bins=50, color='red')  # Histogram for fraud amounts
plt.title("Fraud Transaction Amount Distribution")          # Title for fraud plot
plt.xlabel("Amount ()")                                     # X-axis label
plt.ylabel("Number of Fraud Transactions")                  # Y-axis label
plt.show()
```

```python
# Step 7: Correlation Heatmap

# Create a figure of size 12x8 inches
plt.figure(figsize=(12,8))

# Compute the correlation matrix of the dataset
corr_matrix = df.corr()

# Plot the heatmap of correlations
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)  # 'coolwarm' color map, lines between cells

# Add title to the heatmap
plt.title("Correlation Heatmap")

# Display the heatmap
plt.show()
```

```python
# Step 8: Correlation with Target (Class)

# Extract the correlation of each feature with the target column 'Class'
corr_with_class = corr_matrix['Class'].sort_values(ascending=False)  # Sort from highest to lowest correlation

# Print the correlations to see which features are most positively or negatively associated with fraud
print("Correlation of Features with Fraud Class:\n", corr_with_class)
```

```python
# Step 9: Distribution of PCA Features (V1V28)

# Create a list of PCA feature column names V1 to V28
pca_features = [f"V{i}" for i in range(1, 29)]

# Create a large figure to accommodate multiple subplots
plt.figure(figsize=(20,20))

# Loop through each PCA feature and plot its distribution
for i, feature in enumerate(pca_features, 1):
    plt.subplot(7, 4, i)  # Arrange subplots in a 7x4 grid
    sns.histplot(df[feature], bins=50, color='purple')  # Histogram for each PCA feature
    plt.title(feature)  # Title for each subplot

# Adjust layout so subplots do not overlap
plt.tight_layout()
plt.show()
```
